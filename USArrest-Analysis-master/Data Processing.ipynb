```python
# Misal kita bagi level Murder menjadi 3 kelas (low,medium,high)
k = 3 # banyak level
i = (max(USArrest.Murder)-min(USArrest.Murder))/k # lebar interval
print(k,i)
```

```python
# Membuat Fungsi level murder
def murder_funct(x):
    if x >= 11.8:
        return 'high'
    elif x >= 6.3:
        return 'medium'
    else:
        return 'low'
```

```python
murder_level = USArrest.Murder.apply(murder_funct)
```

```python
murder_level.head(10)
```

```python
USArrest['murder_level'] = murder_level
```
