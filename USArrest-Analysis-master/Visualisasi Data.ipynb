```python
import numpy as np
import pandas as pd
```

```python
USArrest = pd.read_csv('D:/Programming/Data Source/USArrests.csv',sep=',')
```

```python
USArrest.head(10)
```

```python
# basic plot
import matplotlib.pyplot as plt
```

```python
# Memperbolehkan grafik untuk muncul di Jupyter Notebook
%matplotlib inline
```

```python
# seaborn dibangun di atas matloplib
import seaborn as sns
sns.set(font_scale=1.2)
```

```python
#==== Scatterplot UrbanPop vs Murder
plt.scatter('UrbanPop','Murder',data=USArrest,marker='o',)
plt.title('Scatterplot Urban Population vs Murder')
plt.xlabel('Urban Population')
plt.ylabel('Murder')
plt.show()
```

```python
#==== Barplot for total Murder each country
plt.figure(figsize=(8, 8))
# plt.bar('City','Murder',data=USArrest) # Vertical Plot
plt.barh('City','Murder',data=USArrest) # Horizontal Plot
plt.title('Total Murder For Each Country')
plt.xlabel('Country')
plt.ylabel('Murder')
plt.show()
```

```python
#===== Bubble Chart
plt.scatter('Assault','Murder',data=USArrest,s='Assault',alpha=0.5)
plt.title('Assault vs Murder For Each Country')
plt.xlabel('Assault') 
plt.ylabel('Murder')
plt.show()
```
